```python
from ultralytics import YOLO

# Load a model
model = YOLO("yolo11n.pt")  # load a pretrained model (recommended for training)

# Train the model
results = model.train(
    data    = "../configs/yolo11n.yaml", 
    epochs  = 5, 
    imgsz   = 640,
    batch   = 8,                # sesuaikan dengan VRAM GPU
    device  = 0,             # "0" untuk GPU, "cpu" untuk CPU
    amp = True,              # Automatic Mixed Precision (AMP) untuk mempercepat training
    project = "../outputs/models/",       # folder output hasil training
    name    = "ldxray_yolo11n",  # nama subfolder run
    patience= 5,               # early stopping: stop jika tidak improve 10 epoch
    save    = True,             # simpan best & last checkpoint
    # val     = True,             # validasi tiap epoch
    # exist_ok= True,             # overwrite jika folder sudah ada
    workers = 0,                # jumlah CPU workers dataloader
    cache   = "False"           # cache gambar ke RAM (True jika RAM besar)
    )
```

```python
yolo task=detect mode=train model=yolo11n.pt data=../configs/yolo11n.yaml epochs=5 batch=16 workers=8 device=0 project=../outputs/models/ name=ldxray_yolo11n patience=5 save=True
```
